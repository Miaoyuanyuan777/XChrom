# XChrom Quick Start Guide

This notebook provides a quick start example for training XChrom tool. 
XChrom is designed for chromatin accessibility prediction using deep learning models. The notebook demonstrates how to load example data, train a model, and visualize the training results.

### Overview
- **Purpose:**                       
  To help users quickly verify that XChrom is installed and functioning correctly by running a small-scale test.
- **Workflow:**
    1. Import the XChrom package.
    2. List and access the example data provided for testing.
    3. Train the XChrom model using the example data.
    4. Visualize and save the training history.

### Data
- The test data consists of a small subset (100 cells and 1000 peaks) sampled from the original dataset.
- Preprocessed training and test data are provided for convenience.
- Data files are located in the 'xchrom/data/train_data/' directory.
- The raw cellembedding h5ad file is located in the 'xchrom/data/test_rna.h5ad' directory.

### Output
- The trained model is saved in the './data/quick_start/E1000best_model.h5'.
- The training history is saved in the './data/quick_start/history.pickle', and './data/quick_start/train_history_plot.pdf'.

**Note:** This example is for demonstration and testing purposes only. The dataset is small and not intended for research conclusions.


```python
import xchrom as xc
```

```python
# list all available files and directories in xchrom test data
print("available files and directories:", xc.list_items())
# get data directory path
print("data directory:", xc.get_data_dir())
```

```python
data_path = xc.get_data_dir()
```

```python
history = xc.tr.train_XChrom(
    input_folder = f'{data_path}/train_data',
    cell_embedding_ad = f'{data_path}/test_rna.h5ad',
    cellembed_raw='X_pca',
    out_path='./data/quick_start/train_out/',
    epochs = 10,
    verbose = 1
)
```

```python
xc.pl.plot_train_history(
    history = history['history'],
    savefig = True,
    out_file = './data/quick_start/train_out/train_history_plot.pdf'
    )
```

XChrom has been successfully installed and loaded! You can now use XChrom for analysis.
