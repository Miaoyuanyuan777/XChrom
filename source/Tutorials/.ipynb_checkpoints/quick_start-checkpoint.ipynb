## 在这里进行模型训练，测试XChrom工具是否可以正常使用
- 原始数据链接在 , 我们从中随机取样了100个cells和1000个peaks作为总数据进行测试，我们进行了数据的前处理，生成了train和test数据
- 测试数据可以从链接   下载, 这里只是一个很小的用来测试的数据, 本身不具有研究价值,包括atac预处理后生成的训练数据, 以及提供raw cellembedding的h5ad文件
- 下载后的训练数据可以保存在: './data/quick_start/train_data/', 提供raw cellembedding的h5ad文件保存在: './data/quick_start/test_rna.h5ad'

```python
import xchrom as xc
```

```python
history = xc.tr.train_XChrom(
    input_folder='./data/quick_start/train_data/',
    cell_embedding_ad='./data/quick_start/test_rna.h5ad',
    cellembed_raw='X_pca',
    out_path='./data/quick_start/train_out/',
    epochs = 10,
    verbose = 1
)
```

```python
xc.pl.plot_train_history(
    history = history['history'],
    savefig = True,
    out_file = './data/quick_start/train_out/train_history_plot.pdf'
    )
```
